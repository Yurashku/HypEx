```python
import cProfile
import pstats
import sys
import io

from hypex import ABTest
from hypex.dataset import Dataset, InfoRole, TargetRole, TreatmentRole
```

```python
pr = cProfile.Profile()
pr.enable()
```

```python
data = Dataset(
    roles={
        "user_id": InfoRole(int),
        "treat": TreatmentRole(),
        "pre_spends": TargetRole(),
        "post_spends": TargetRole(),
        "gender": TargetRole()
    }, data="data.csv",
)
data
```

```python
test = ABTest()
result = test.execute(data)
```

```python
result.resume
```

```python
pr.disable()
pr.dump_stats("out.prof")
```

```python
s = io.StringIO()
ps = pstats.Stats(pr, stream=s).sort_stats("cumulative")
ps.print_stats("hypex", 20)

print("top 20 functions by cumulative time:")
print(s.getvalue())
```
